## Get String Lengths

```python
def string_length(column):
    length = []
    for index in neuro_df[column]:
        l1 = (len(str(index)))
        length.append(l1)
    print(f"{column}: {max(length)}")
```

```python
string_length("title")
string_length("citation_author")
string_length("citation_journal_title")
string_length("description")
string_length("keywords")
```
